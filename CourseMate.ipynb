```python
# Columns to check for non-null values
required_columns = ["Title", "URL", "Short Intro", "Category", "Sub-Category", "Course Type", "Language", "Skills"]

# Drop rows with NaN in any of the required columns
filtered_courses = course.dropna(subset=required_columns)

# Check the filtered result
print(filtered_courses.info())  # To see data structure
print(filtered_courses.head())
```

```python
# Load the dataset yang sudah difilter sebelumnya (dengan 1960 rows)
course_data = filtered_courses.copy()

# Fungsi untuk membersihkan teks
def clean_text(text):
    if isinstance(text, str):
        text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove punctuation
        text = re.sub(r'\s+', ' ', text)         # Remove extra spaces
    return text

# Bersihkan kolom 'Skills' dan 'Short Intro'
course_data['Skills'] = course_data['Skills'].apply(clean_text)
course_data['Short Intro'] = course_data['Short Intro'].apply(clean_text)
```

```python
# Melihat semua kategori unik yang ada di dataset
unique_categories = course_data['Category'].unique()
print("Kategori yang tersedia dalam dataset:")
print(unique_categories)
```

```python
# Melihat semua skill unik yang ada di dataset
unique_skills = course_data['Skills'].unique()
print("Skill yang tersedia dalam dataset:")
print(unique_skills)
```

```python
# Memisahkan setiap keterampilan yang ada dalam dataset dan membersihkan duplikat
def clean_and_separate_skills(skill_string):
    # Pisahkan keterampilan berdasarkan spasi
    skills = skill_string.split()

    # Bersihkan setiap keterampilan (misalnya jika ada spasi tambahan atau karakter lain)
    cleaned_skills = [skill.strip().lower() for skill in skills if skill.strip()]

    return cleaned_skills

# Gabungkan semua keterampilan dalam dataset menjadi satu list
all_skills = []

# Menggunakan apply untuk setiap baris dalam kolom 'Skills' dan menambahkannya ke all_skills
for skill_string in course_data['Skills']:
    all_skills.extend(clean_and_separate_skills(skill_string))

# Menghilangkan duplikat keterampilan
unique_skills = list(set(all_skills))

# Menampilkan keterampilan yang telah dibersihkan dan terpisah
print(f"Jumlah keterampilan unik yang telah dibersihkan: {len(unique_skills)}")
print(unique_skills[:20])  # Menampilkan 20 keterampilan pertama
```

```python
# Fungsi rekomendasi berbasis kategori dan keterampilan
def recommend_courses(user_category, data, top_n=5):
    # Filter data berdasarkan kategori yang dipilih
    category_filtered = data[data['Category'].str.lower() == user_category.lower()].copy()

    return category_filtered[['Title', 'Short Intro', 'URL', 'Course Type']]
```

```python
# Membuat widget input kategori
category_widget = widgets.Dropdown(
    options=course_data['Category'].unique(),
    description='Category:',
    disabled=False
)

# Fungsi untuk menampilkan hasil rekomendasi
def show_recommendations(user_category):
    # Mengambil data rekomendasi berdasarkan kategori
    recommended_courses = recommend_courses(user_category, course_data)

    # Menampilkan hanya 20 rekomendasi pertama
    recommended_courses = recommended_courses.head(15)

    # Menampilkan hasil
    print(f"Recommended Courses for Category: {user_category}\n")
    for idx, row in recommended_courses.iterrows():
        print(f"Title: {row['Title']}")
        print(f"Intro: {row['Short Intro']}")
        print(f"URL: {row['URL']}")
        print(f"Course Type: {row['Course Type']}\n")
        print("=" * 20)

# Menampilkan form input interaktif
widgets.interactive(show_recommendations,
                    user_category=category_widget)
```
